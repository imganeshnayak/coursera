```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(boston_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x='CHAS', data=boston_df)
plt.title("Charles River Distribution")
plt.show()
```

```python
boston_df['AGE_GROUP'] = pd.cut(
    boston_df['AGE'],
    bins=[0,35,70,100],
    labels=['New','Mid','Old']
)
```

```python
sns.boxplot(x='AGE_GROUP', y='MEDV', data=boston_df)
plt.title("MEDV vs Age Group")
plt.show()
```

```python
sns.histplot(boston_df['PTRATIO'], bins=20)
plt.title("Distribution of Pupil Teacher Ratio")
plt.show()
```
